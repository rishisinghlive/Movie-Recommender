```python
import ast
```

```python
# This will remove the ' ' from every list of dictionary
ast.literal_eval(df['genres'].iloc[0])
```

```python
# create a function which will extract the name 
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
# Apply the above function over the features
df['genres'] = df['genres'].apply(convert)
```

```python
ast.literal_eval(df['crew'][0])
```

```python
# Will looking for job == Director and take the name
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df
```

```python
# Converting in a list of string
df['overview'] = df['overview'].apply(lambda x:x.split())
```

```python
# Removeing White space
df['genres'] = df['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
df['production_companies'] = df['production_companies'].apply(lambda x:[i.replace(' ', '') for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
df
```

```python
# Merge all the list of keywords into one
df['tag'] = df['overview'] + df['genres'] + df['keywords'] + df['production_companies'] + df['cast'] + df['crew']
```

```python
movie = df[['movie_id', 'title', 'tag']]
```

```python
# removing the list
movie['tag'] = movie['tag'].apply(lambda x:' '.join(x))
```

```python
# make Lower case
movie['tag'] = movie['tag'].apply(lambda x:x.lower())
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
movie['tag'] = movie['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# will convert the corpus into a 6000 dimension vector
cv = CountVectorizer(max_features=6000, stop_words='english')
```

```python
vector = cv.fit_transform(movie['tag']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
# Will calculate the Cosine Similarity between the each movie corpus
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
movie
```

```python
# Will take the index of heighest 5 similarity scorer
sorted(list(enumerate(similarity[1])), reverse=True, key= lambda x:x[1])[1:6]
```

```python
# Create a function that will fetch the search movie index and based of similarity score it returns top 5 movies
def recomend(keyword):
    movie_index = movie[movie['title'] == keyword].index[0] # Find the index of the searched movie
    distances = similarity[movie_index] # Use index and Finds the similarity score of that movie
    
    # enumrate the similarity score to keep the index value intact and fetch the top 5 similar movie indexes
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6] 
    
    for i in movies_list: # Will take the index of the movie and fetch the movie name
        print(movie.iloc[i[0]].title)
```

```python
recomend('Krrish')
```
